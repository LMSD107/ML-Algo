**Using Entropy**

```python
import numpy as np
import pandas as pd
from sklearn import metrics
```

```python
df=pd.read_csv("/content/course.csv")
value=['CAT1','CAT2','FAT','Internal']
df
```

```python
from sklearn import preprocessing
string_to_int= preprocessing.LabelEncoder()                
df=df.apply(string_to_int.fit_transform) 
df
```

```python
feature_cols = ['CAT1','CAT2','FAT','Internal']
X = df[feature_cols]
y = df.Grade
```

```python
X_train=X.iloc[0:10]
y_train=y.iloc[0:10]
X_test=X.iloc[10:11]
y_test=y.iloc[10:11]
```

```python
from sklearn.tree import DecisionTreeClassifier                            
classifier =DecisionTreeClassifier(criterion="entropy", random_state=5)   
classifier.fit(X_train, y_train)  
```

```python
y_pred= classifier.predict(X_test)
from sklearn.metrics import accuracy_score
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

data_p=pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})  
data_p
```

```python
from sklearn.metrics import classification_report, confusion_matrix  
print(confusion_matrix(y_test, y_pred))  
print(classification_report(y_test, y_pred))
```

```python
import graphviz
from sklearn.tree import export_graphviz 
import six
import sys
sys.modules['sklearn.externals.six'] = six
from sklearn.externals.six import StringIO
import pydotplus
from IPython.display import Image
dot_data = StringIO()
print(dot_data)
```

```python
export_graphviz(classifier, out_file=dot_data, filled=True, rounded=True, special_characters=True,feature_names=value,class_names=['0','1','2'])
```

```python
graph=pydotplus.graph_from_dot_data(dot_data.getvalue())
```

```python
Image(graph.create_png())
```
